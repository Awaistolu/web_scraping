```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.pakwheels.com/used-cars/search/-/mk_honda/ct_lahore/?page=1'
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
response = requests.get(url,headers=headers)
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
all_list=soup.find_all('div', class_='well')
```

```python
all_list[1].find('a',class_='car-name').text.strip()
```

```python
name=[]
price=[]
city=[]
year=[]
milage=[]
engine=[]

for i in range(1,len(all_list)):
  try:
    name.append(all_list[i].find('a',class_='car-name').text.strip())
  except:
    name.append(np.nan)
  try:
    price.append(all_list[i].find('div',class_='price-details').text.strip())
  except:
    price.append(np.nan)
  try:
    city.append(all_list[i].find_all('ul',class_='list-unstyled')[2].text.strip())
  except:
    city.append(np.nan)
  try:
    year.append(all_list[i].find_all('li')[0].text.strip())
  except:
    year.append(np.nan)
  try:
    milage.append(all_list[i].find_all('li')[1].text.strip())
  except:
    milage.append(np.nan)
  try:
    engine.append(all_list[i].find_all('li')[2].text.strip())
  except:
    engine.append(np.nan)
d={'Name':name ,'Price':price ,'City':city ,'Year':year ,'Milage':milage ,'Engine':engine }
df=pd.DataFrame(d)
```
