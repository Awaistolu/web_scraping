```python
new_df =pd.DataFrame()
for j in range(1,65):
  url='https://www.yellowpages.ca/search/si/{}/Grocery+Stores/Toronto+ON'.format(j)
  headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
  response=requests.get(url,headers=headers)
  soup = BeautifulSoup(response.content,'html.parser')
  all_list = soup.find_all('div',class_='listing__content__wrapper')
  name=[]
  address=[]
  phone_no=[]


  for i in range(1,len(all_list)):
    try:
      name.append(all_list[i].find('h3').text.strip())
    except:
      name.append(np.nan)
    try:
      address.append(all_list[i].find('span',class_='listing__address--full').text.strip())
    except:
      address.append(np.nan)
    try:
      phone_no.append(all_list[i].find_all('li',class_='mlr__item mlr__item--more mlr__item--phone jsMapBubblePhone')[0].text.strip())
    except:
      phone_no.append(np.nan)
  d={'Name':name,'Address':address,'Phone_no':phone_no}
  df=pd.DataFrame(d)
  new_df = new_df.append(df, ignore_index=True)
```
